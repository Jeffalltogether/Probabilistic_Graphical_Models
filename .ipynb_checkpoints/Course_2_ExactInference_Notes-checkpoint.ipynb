### 1.1 Important Distinctions  
There are three main decisions that are selected when choosing our representation structure:  

* Template V. Specific  
* Directed V. Undirected  
* Generative V. Discriminative  
* **Hybrids are common**, for instance a model with a template component and a specific component  

### 1.2 Template V. Specific   
Examples of template and specific models

| Template           |   Hybrids          |   Specific         |
|--------------------|--------------------|--------------------|
|Image segmentation  | fault diagnosis    | medical diagnosis  |


**Characteristics**  
Template:
* Small number of variable types 
* Effort goes into finding the most predictive features over complicated model design
* More on feature engineering

Specific:
* Large number of unique variables  
* Each variable requires it's own model  
* More about model design  
 
### 1.3 Generative V. Discriminative  

**Characteristics**  
Generative:  
* When there is more than one task or no pre-determined task (e.g., in medical diagnosis when we may want to know the probability of a variety of diseases given a subset of symptoms)  
* Easier to train in certain regimes (e.g., when there is missing data)  

Discriminative:  
* Used when we have a particular prediction task in mind  
* We will use richly discriminative features in this model  
* Modeling as a discriminative task **allows us to avoid dealing with correlations between features**   
* Usually higher performance in this specific task  

For which of the following applications might a generative model be a better choice than a discriminative model?  

Answer:   
Medical diagnosis where obtaining ground-truth labels for the data is expensive.  
and,  
Modeling different music composers for classification of music and creating new music pieces.  

### 1.4 Variable Types  

* Target $\leftarrow:$ outcome variable, $y$, disease we wish to predict  
* Observed $\leftarrow:$ input variables, $x_n$, symptoms of the patient  
    * including complex, constructed (engineered) features  
* Latent $\leftarrow:$ hidden, variables that we don't care to predict, **simplify the structure of the model**  

#### Example of Latent Variable
If we wanted to predict the current time using the time present on the watches from all the students in the classroom, we know that all of these watches are correlated.  One way to represent this is to connect each watch in the classroom by a vertex in our model.  However, this generates a very complex fully-connected model.  A better way is to assume all watches are are correlated to Greenwich Mean Time (GMT), and connect each watch to this **latent variable**. ### 1.5 Structure 

When would we consider a directed or non-directed model?

* Causal vs. Non-causal ordering  
In a directed graph:  
In the case of modeling the independencies between $X$ and $Y$, we can model all scenarios like:  
$X \rightarrow Y$, or  
$X \leftarrow Y$  

However, in the case of three variables, where we start with the graph $G = X_1 \rightarrow Y \leftarrow X_2$. The independency modeled here is $I(G) = (X_1 \perp X_2)$.  If we want to reverse the arrow between $Y$ and $X_2$ to generate this structure $X_1 \rightarrow Y \rightarrow X_2$ the independency in this new graph, $G'$, is $I(G') = (X_1 \perp X_2 \mid Y)$.  

Unfortunately, the independencies in $G'$, specifically $(X_1 \perp X_2 \mid Y)$, are not found in the original model  $G$. In other words, $G'$ is not an *I-map* of $G$.   

In order for the new model $G'$ to have this relationship $Y \rightarrow X_2$, and maintain an independence structure that is contained in the original graph, we need to place a connection $X_1 \rightarrow X_2$.  The I-map for this new graph is $\emptyset$ and is therefore $\subseteq G$.  

In the watch example, this is what would happen if we were asked to reverse the direction of the connections between the watches and the latent variable GMT:  

<img src="./images/notes_1_2.png">  

The problem here is that we need to add the red-lines to maintain that all the watches are correlated to one another, resulting in the fully connected graph we were trying to avoid (by adding the latent variable) in the first place.  

In summary...  

**Causal Ordering is:**  
* Sparser (fewer variables or connections needed)  
* more intuitive  
* easier to parameterize  

### 1.6 Constructing a graph using a practical technique

A practical modeling scenario:  

The conversation between the modeler and the domain expert (doctor) can go as follows to generate a model:

1. begin with the outcome variable and ask the domain expert, "what variables influence whether we have this outcome (e.g., lung cancer)?"
2. The domain expert may offer a couple variables (e.g., smoking and kind of work)
3. then we would ask the same expert or another, what do these variables depend on.
4. We continue to extend the conversation backwards until we reach factors that are easily determined, such as P(age) or P(gender), or P(country of origin). At this point we can stop going in this backwards direction.
5. Now we can look in the other direction for factors that would indicate the presence or absence of the outcome variable we want to predict.  We would ask the expert for variables that are evidence of the outcome variable (e.g., coughing, bloody sputum, lung nodules present on CT scan, or positive biopsy are indicators of lung cancer).  
6. we might then ask, "what is the probability of a patient having cancer given a coughing?" and the expert may say that it would depend on other variables, so we can continue extending the model in the evidencial direction until we reach simple to ascertain variables.

### 1.7 Parameters: Values
What matters:
* **Zeros** have a large effect on the model (things shouldn't have a probability of 0 unless it's definitial)
* Orders of magnitude (1/10 vs. 1/100)
* Relative values

Making the attire of people observed variables and the weather states target variables.
